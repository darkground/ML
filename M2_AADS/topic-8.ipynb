## Итоговый тест

### Вопрос 1. Зависимость скорости работы алгоритма от числа n описывается как `50*n*log(n) + n^2 + 100n + 10000`. Какие из следующих асимптотический оценок включают эту функцию?

- Θ(n*log(n) - 100)
- Ω(n) ✅
- Ο(n^2) ✅
- Ο(50*n*log(n))

### Вопрос 2. Рекуррентное соотношение имеет вид: T(n) = 2T(n/2) + Ο(1). Какой из перечисленных алгоритмов имеет такую временную оценку?
- Быстрая сортировка
- Двоичный поиск
- Обход двоичного дерева ✅

### Вопрос 3. Какая из следующих сортировок всегда работает за Ο(n^2)?
- Выборкой (selection) ✅
- Пузырьком (bubble)
- Быстрая сортировка (quicksort)
- Вставками (insertion)

### Вопрос 4. Какая из следующих эффективных сортировок требует использования дополнительной памяти?
- Быстрая сортировка (quicksort)
- Сортировка слияниями (mergesort) ✅

### Вопрос 5. Динамический массив — это реализация списка, которая требует линейное время при добавлении элемента:
- в конец
- в середину ✅
- в начало ✅

### Вопрос 6. Временная сложность поиска в двоичном дереве поиска имеет следующую оценку в худшем случае:
- Ο(1)
- Ο(n) ✅

### Вопрос 7. Минимальная высота АВЛ-дерева:

### Вопрос 8. Какие из перечисленных структур данных относятся к линейным:

- Матрица

### Вопрос 10. АВЛ-дерево может быть использовано как реализация ассоциативного массива. Какие операции в этом случае имеют оценку сложности Ο(log(n))?
- Сортировка
- Вставка ✅
- Удаление ✅
- Поиск ✅

### Вопрос 11. С чем связан факт, что быстрая сортировка в худшем случае работает за Ο(n^2).
- Эта сортировка всегда работает с этой оценкой
- Неправильный выбор опорного элемента ✅
- Сортировка использует комбинацию других сортировок в качестве работы

### Вопрос 12. Какой из приведённых случаев является лучшим для сортировки вставками?
- Последовательность содержит только одинаковые элементы ✅
- Последовательность элементов не отсортирована
- Такого случая нет
