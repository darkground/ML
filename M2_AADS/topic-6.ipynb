## Тест 7. B-деревья

### Вопрос 1. Пусть m — максимальное число дочерних узлов в B-дереве. Тогда верно, что минимальное число элементов для любого некорневого элемента равно:

- 1
- log m
- m
- m / 2 ✅
- 2

### Вопрос 2. Пусть m — максимальное число дочерних узлов в B-дереве. Тогда верно, что минимальное число элементов для корневого элемента равно:

- m
- 2 ✅
- log m
- m / 2

### Вопрос 3. Как связано количество доступных ключей, хранящихся в одном узле, и количество дочерних узлов, если количество ключей k, а дочерних узлов m?

- k = log m
- k = m / 2
- k = m - 1 ✅

### Вопрос 4. B-дерево можно называть:

- Двоичным
- Деревом поиска ✅
- Сбалансированным ✅

### Вопрос 5. Как высота поддеревьев одного узла может отличаться?

- В 2 раза
- Никак, высота одинаковая ✅
- На 1

### Вопрос 6. При вставке нового элемента в этом случае будет выполнена операция разделения:

- Когда количество дочерних элементов максимально ✅
- Когда количество дочерних элементов равно 1
- Когда количество дочерних элементов минимально

### Вопрос 7. До какого момента происходит вытеснение элемента выше при операции разделения?

- До тех пор, пока вытесненный элемент делает родительный узел переполненным ✅
- До тех пор, пока вытесненный элемент не становится новым корнем дерева ✅
- Один раз в верхний узел
