Двоичное дерево писка состоит из узлов, где один узел содержит 3 поля:
- Данные
- Ссылка на левый потомок
- Ссылка на правый потомок

Данная структура может быть представлена через словари, например, следующим образом:

```py
t = {
    "key": 2,
    "left": {
        "key": 1
    },
    "right": {
        "key": 3
    }
}
```

Реализуйте процедуру поиска в двоичном дереве поиска, исходя из его основного свойства, что элементы слева должны быть меньше, а элементы справа — больше.
```py
def find(tree, key)
```

Для приведённого выше примера `find(t, 1)` вернёт ссылку на узел
```py
{'key': 1}
```
Если элемент не найден в дереве, то процедура должна возвращать None

````python
def find(tree, key):
    if tree is None:
        return None
    if tree['key'] > key:
        return find(tree.get('left'), key)
    elif tree['key'] < key:
        return find(tree.get('right'), key)
    else:
        return tree

# Solution Test

t = {
    "key": 2,
    "left": {
        "key": 1
    },
    "right": {
        "key": 3
    }
}

print(find(t, 2))
````

Реализуйте обход дерева в глубину по схеме: левый узел - корень - правый узел на двоичном дереве поиска. Такой обход должен порождать отсортированные последовательности, например:

```py
dfs({
    "key": 2,
    "left": {
        "key": 1
    },
    "right": {
        "key": 3
    }
})
```

Печатает в консоль следующие значения
```
1
2
3
```

````python
def dfs(tree):
    if tree is None:
        return
    dfs(tree.get('left'))
    print(tree['key'])
    dfs(tree.get('right'))

# Solution Test

t = {
    "key": 2,
    "left": {
        "key": 1
    },
    "right": {
        "key": 3
    }
}

dfs(t)
````

Реализуйте обход дерева в ширину. Такой обход должен порождать последовательности значений как при обходе дерева сверху вниз, слева направо, т.е.

```py
bfs({
    "key": 2,
    "left": {
        "key": 1
    },
    "right": {
        "key": 3
    }
})
```

печатает в консоль следующие значения
```
2
1
3
```

````python
def bfs(tree):
    queue = [tree]
    while len(queue) > 0:
        elem = queue.pop(0)
        print(elem['key'])
        if 'left' in elem:
            queue.append(elem['left'])
        if 'right' in elem:
            queue.append(elem['right'])

# Solution Test

bfs({
    "key": 2,
    "left": {
        "key": 1
    },
    "right": {
        "key": 3
    }
})
````
