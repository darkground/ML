## Тест 5. Задачи сортировки

### Вопрос 1. Каким образом можно получить доступ к данных в ассоциативном массиве:

- По ключу можно получить значение ✅
- По индексу можно получить элемент
- Никак, только обойти всю структуру

### Вопрос 2. Какое ограничение накладывают на ассоциативный массив:

- Он не может использовать None в качестве ключа
- Он не может содержать пустые строки в качестве ключа
- Он не может использовать дубликаты ключей ✅

### Вопрос 3. Какие реализации можно использовать для ассоциативного массива?

- Использовать сбалансированные двоичные деревья ✅
- Хранить как список пар «ключ-значение» ✅
- Использовать хеш-таблицы ✅

### Вопрос 4. Какая оценка времени работы поиска значения по ключу для сбалансированного двоичного дерева поиска?

- O(n^2)
- O(n)
- O(log n) ✅

### Вопрос 5. Какое ограничение накладывают на ключи, чтобы их можно было использовать в сбалансированном двоичном дереве поиска?

- Они не должны отрицательным значением
- Они должны быть сравнимы между собой ✅
- Оно должны быть не 0 и не пустыми

### Вопрос 6. Что такое хеш-функция?

- Функция, которая отображает данные сама на себя
- Функция, которая преобразует массив данных произвольной длины в выходную строку установленной длины ✅
- Функция, которую можно изобразить на графике с помощью логарифмических операций

### Вопрос 7. Хеш-функции должны быть:

- Иметь минимальное число коллизий ✅
- Быстро вычислимы ✅

### Вопрос 8. Что такое коллизия?

- Ситуация, когда функция от некоторого аргумента равна этому аргументу
- Ситуация, при которой значение некоторое хеш-функции для двух разных значений равны ✅
- Ситуация, при которой невозможно вычислить аргумент по заданному значению функции

### Вопрос 9. Какие существуют методы борьбы с коллизиями?

- Метод открытой адресации ✅
- Метод цепочек ✅
- Метод закрытой адресации

### Вопрос 10. Метод цепочек в случае коллизии:

- Удаляет предыдущее значение
- Хранит значения, хеши которых совпали, в связном списке ✅
- Перераспределяет элементы в массиве

### Вопрос 11. Метод открытой адресации в случае коллизии:

- Хранит значения, хеши которых совпали, в связном списке
- Удаляет предыдущее значение
- Перераспределяет элементы в массиве ✅

### Вопрос 12. Какая сложность вставки в хеш-таблицу в лучшем/среднем случае?

- O(n)
- O(log n)
- O(1) ✅

### Вопрос 13. Какая сложность вставки в хеш-таблицу в худшем случае?

- O(1)
- O(log n)
- O(n) ✅

### Вопрос 14. Какая сложность получение элемента по ключу в хеш-таблице в лучшем/среднем случае?

- O(1) ✅
- O(log n)
- O(n)

### Вопрос 15. Какая сложность получение элемента по ключу в хеш-таблице в худшем случае?

- O(1)
- O(n) ✅
- O(log n)

### Вопрос 16. Какая сложность добавление элемента в сбалансированное двоичное дерево поиска в среднем?

- O(1)
- O(n)
- O(log n) ✅

### Вопрос 17. Какая сложность получения элемента в сбалансированное двоичное дерево поиска в худшем случае?

- O(n)
- O(1)
- O(log n) ✅

### Вопрос 18. С чем связан тот факт, что время поиска элемента в хеш-таблице различаются в среднем и худшем случаях?

- Наличие дубликатов
- Наличие None
- Наличие коллизий ✅
