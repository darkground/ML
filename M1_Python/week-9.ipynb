```python
# Task 4-1-1-1
# На вход программе через input подается строка, состоящая из символов алфавита и цифр.
# Выведите на экран True, если количество символов равно количеству цифр в строке. Иначе выведите False.
# Использование функционального программирования приветствуется.

inp = input()
digits = len([x for x in inp if x.isdigit()])
nondigits = len(inp) - digits
print(digits == nondigits)
```

```python
# Task 4-1-1-2
# На вход подается строка, состоящая из символов алфавита и цифр.
# Замените каждый третий символ строки на символ "-" (начиная с нуля).
# Выведите получившуюся строку на экран. Вы можете перейти к списку символов,
# чтобы было проще заменять элементы. Но не забудьте потом перейти снова к строке.
# Использование функционального программирования приветствуется. 

inp = input()
out = ""
for i, e in enumerate(inp):
    out += '-' if i % 3 == 0 else e
print(out)
```

```python
# Task 4-1-1-3
# На вход программе через input подается строка, состоящая из символов алфавита,
# цифр и других служебных символов. Создайте новую строку, оставив только символы алфавита и цифры.
# Символы алфавита в строке должны быть приведены к верхнему регистру.
# Порядок элементов менять нельзя. Выведите получившийся список на экран.
# Использование функционального программирования приветствуется. 

inp = input()
out = "".join(filter(lambda x: x.isalpha() or x.isdigit(), inp)).upper()
print(out)
```

```python
# Task 4-1-1-4
# На вход программе через input подается строка, в которой записаны элементы списка
# -- строки, записанные через пробел. Строки состоят из символов английского алфавита.
# Перейдите от строки к списку элементов. Порядок элементов менять нельзя.
# Создайте новый список и добавьте туда только те элементы-строки, в составе которых
# есть не менее трех гласных (вне зависимости от регистра).
# Гласные английского алфавита: a, e, i, o, u, y. Выведите итоговый список на экран.
# Использование функционального программирования приветствуется. 

def has3vowels(e):
    return len([x for x in e if x.lower() in "aeiouy"]) >= 3

inp = input()
lst = inp.split()
fil = [x for x in lst if has3vowels(x)]
print(fil)
```

```python
# Task 4-1-1-5
# На вход программе через input подается строка, в которой записаны элементы списка
# -- целые числа (положительные и отрицательные), записанные через пробел.
# Перейдите от строки к списку целых чисел. Порядок элементов менять нельзя.
# Посчитайте частоту встречаемости каждого элемента списка.
# Создайте новый список и только тех элементов исходного списка,
# которые встречаются более 2 раз. Выведите итоговый список на экран.
# В итоговом списке должны быть объекты типа int.
# Использование функционального программирования приветствуется. 

inp = input()
lst = list(map(int, inp.split()))

occur = {}
for x in lst:
    occur.setdefault(x, 0)
    occur[x] += 1

fil = [x for x in lst if occur[x] > 2]
print(fil)
```

```python
# Task 4-1-1-6
# На вход подается строка, в которой записаны элементы списка
# -- целые числа через пробел. Перейдите от строки к списку объектов типа int.
# Создайте новый список и добавьте в него только те числа из исходного списка,
# которые делятся без остатка на хотя бы одно другое число из этого же списка.
# Выведите новый список на экран. В итоговом списке должны быть объекты типа int.
# Использование функционального программирования приветствуется.

def divides(e, st):
    return any([e % x == 0 for x in st if x != e])

inp = input()
lst = list(map(int, inp.split()))
uniq = set(lst)
fil = [x for x in lst if divides(x, uniq)]
print(fil)
```

```python
# Task 4-1-1-7
# Напишите функцию solve(), которая принимает на вход список целых чисел и
# возвращает список значений квадратных корней из четных чисел исходного списка.
# Для извлечения корня воспользуйтесь функцией sqrt из модуля math (в списке только положительные числа).
# Результат округлите до 2-х знаков после запятой с помощью round(value, 2). При решении обратите внимание, что:
#   - Функция не должна ничего выводить на экран.
#   - Кроме определения функции ничего другого писать не нужно.
#   - Использование функционального программирования приветствуется.

from math import sqrt

def solve(L):
    return [round(sqrt(x), 2) for x in L if x % 2 == 0]

# Solution Test

print(solve([52, 8, 51, 66, 42]))
```

```python
# Task 4-1-1-8
# Используя функциональное программирование решите следующую задачу.
# Напишите функцию solve(obj), которая принимает на вход список строк и возвращает список,
# в котором содержатся только те элементы исходного списка,
# которые начинаются с прописной буквы и не включают цифр. При решении обратите внимание, что:
#   - Функция не должна ничего выводить на экран.
#   - Кроме определения функции ничего другого писать не нужно.
#   - Использование функционального программирования приветствуется.

def capitalized(e):
    return e[0].lower() != e[0]
    
def hasdigits(e):
    return any([x for x in e if x.isdigit()])

def solve(obj):
    return list(filter(lambda x: capitalized(x) and not hasdigits(x), obj))

# Solution Test

print(solve(['ADSiR', 'HPfAm', 'xd4fcFH', 'SxyIJd']))
```

```python
# Task 4-1-1-9
# Напишите фукнцию check_inheritance(), которая принимает на вход один список:
# наименования некоторых классов и последним элементом в списке идет название некоторого класса.
# Гарантируется, что некоторый класс (последний элемент в списке, переданном на вход),
# является наследником одного или нескольких классов из списка.
# Функция должна вернуть список всех родителей некоторого класса. Используйте функцию issubclass().

def check_inheritance(L):
    last = L.pop()
    return [cl for cl in L if issubclass(last, cl)]
    
# Solution Test

print(check_inheritance([int, str, complex, dict, bool]))
```

```python
# Task 4-1-1-10
# Напишите функцию check_instance(), которая принимает на вход один список.
# В списке два элемента, первый -- список, где перечислены наименования классов.
# Второй элемент -- некоторый объект. Функция возвращает новый список, куда собраны все наименования классов,
# сущностью которых является этот объект. Используйте функцию isinstance().

def check_instance(L): 
    lst, obj = L
    return [cl for cl in lst if isinstance(obj, cl)]

# Solution Test

print(check_instance([[int, str, complex, dict, bool], True]))
```
