```python
# Task 3-6-1-4
# Напишите функцию hsum_array(), на вход которой передается матрица ndarray, состоящая из целых чисел.
# Функция возвращает ndarray-вектор, содержащий значения сумм элементов по строкам.

import numpy as np

def hsum_array(nd: np.ndarray):
    return nd.sum(axis=1)

# Solution Test

a = np.array([[1, 2, 3], [4, 5, 6]])
r = hsum_array(a)
print(a, r)
```

```python
# Task 3-6-1-5
# Напишите функцию vsum_array(), на вход которой передается матрица ndarray,
# состоящая из целых чисел. Функция возвращает ndarray-вектор,
# одержащий значения сумм элементов по столбцам.

import numpy as np

def vsum_array(nd: np.ndarray):
    return nd.sum(axis=0)

# Solution Test

a = np.array([[1, 2, 3], [4, 5, 6]])
r = vsum_array(a)
print(a, r)
```

```python
# Task 3-6-1-7
# Напишите функцию max_array(), на вход которой передается матрица ndarray,
# состоящая из целых чисел. Функция возвращает максимальный элемент матрицы.

import numpy as np

def max_array(nd: np.ndarray):
    return nd.max()

# Solution Test

a = np.array([[1, 2, 3], [4, 5, 6]])
r = max_array(a)
print(a, r)
```

```python
# Task 3-6-1-8
# Напишите функцию mean_array(), на вход которой передается матрица ndarray,
# состоящая из целых чисел. Функция возвращает среднее значение по матрице,
# округленное до 2-х знаков после запятой с помощью функции round(value , 2). 

import numpy as np

def mean_array(nd: np.ndarray):
    return round(nd.mean(), 2)

# Solution Test

a = np.array([[1, 2, 3], [4, 5, 7]])
r = mean_array(a)
print(a, r)
```

```python
# Task 3-6-1-9
# Напишите функцию cos_array(), на вход которой передается матрица ndarray,
# состоящая из значений угла в градусах (0 до 360). Функция возвращает матрицу nddarray,
# состаящую из значений косинуса (numpy.cos). Каждый элемент матрицы нужно округлить до 2-го
# знака после запятой с помощью функции round(value, 2). Для обхода матрицы при округлении
# можно использовать вложенные циклы, размер матрицы можно узнать через поле shape у объекта ndarray. 

import numpy as np

def cos_array(nd: np.ndarray):    
    xl = []
    for i in nd:
        yl = []
        for j in i:
            result = round(np.cos(j), 2)
            yl.append(result)
        xl.append(yl)
    return np.array(xl)

# Solution Test

a = np.array([[213, 105, 194],[239, 26, 149],[242, 231, 189]])
r = cos_array(a)
print(a)
print(r)
```

```python
# Task 3-6-1-10
# Напишите функцию cov_array(), на вход которой передается двумерный вектор ndarray,
# состоящий из целых чисел. Функция возвращает ковариационную матрицу nddarray (numpy.cov).
# Каждый элемент матрицы нужно округлить до 2-го знака после запятой с помощью функции round(value, 2).
# Для обхода матрицы при округлении можно использовать вложенные циклы,
# размер матрицы можно узнать через поле shape у объекта ndarray. 

import numpy as np

def cov_array(nd: np.ndarray):
    nx = np.cov(nd)
    xl = []
    for i in nx:
        yl = []
        for j in i:
            result = round(j, 2)
            yl.append(result)
        xl.append(yl)
    return np.array(xl)

# Solution Test

a = np.array([[2, -1], [8, -5], [-10, -9]])
r = cov_array(a)
print(a)
print(r)
```
